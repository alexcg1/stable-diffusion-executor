```python
from docarray import Document, DocumentArray
```

```python
from jina import Client
```

```python
flow_uri = 'grpcs://7574c555b8.wolf.jina.ai'
```

```python
client = Client(host=flow_uri)
```

now that our client is connected to our application we can start to use it
We check first the a dry_run is working

```python
client.dry_run()
```

then we actually generate some image :)

```python
docs = client.post(
    on='/',
    inputs=[
        Document(
            text='realistic photo of the linux tux attentind an open source conference in 4k'
        )
    ],
    target_executor= 'diffusion'
)
```

```python
docs[0].matches.plot_image_sprites()
```

```python
fav = 0
```

```python
doc_to_upscale = docs[0].matches[fav]
```

```python
docs = client.post(
    on='/',
    inputs=[doc_to_upscale],  
    target_executor= 'upscaling'
)
```
